```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv(r'C:\Users\DINESH\Dropbox\PC\Downloads\hostelmenu (2).csv')
data
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x,y)
```

```python
z=[[28,1,2]]
g=int(model.predict(z))
g
```

```python
if(g==1):
    print("Idly is the food")
elif(g==2):
    print("Chappati is the food")
elif(g==3):
    print("Uppuma is the food")
elif(g==4):
    print("Biriyani is the food")
elif(g==5):
    print("Pongal is the food")
elif(g==6):
    print("Poori is the food")
else:
    print("Doosa is the food")
    
    
```
